```python
import pandas as pd

gladys = pd.read_csv("./gladys.csv")
```

```python
gladys[:100]
```

```python
gladys.columns
```

```python
result_dict = {}
# first 100 items because Gladys only did the first 100.
for index, row in gladys[:100].iterrows():
    filename = row["Filename(2nd edition "]
    output1 = row["Output.1"]
    reflacx = row["REFLACX"] if pd.notna(row["REFLACX"]) else ""
    eyegaze = row["EyeGaze "] if pd.notna(row["EyeGaze "]) else ""

    if pd.notna(filename):
        result_dict[filename] = [output1, reflacx, eyegaze]
```

```python
result_dict
```

```python
items = []

for filename, values in result_dict.items():
    prediction = values[0].strip()
    reflacx = values[1].strip()
    eyegaze = values[2].strip()

    # Use reflacx as reference if it's not empty, otherwise use eyegaze
    reference = reflacx if reflacx != "" else eyegaze

    item = {
        "id": filename,
        "query": "<image>\nProvide a description of the findings in the radiology image given the following indication: ",
        "reference": reference,
        "prediction": prediction,
    }

    items.append(item)
```

```python
import json

with open("./prediction.jsonl", "w") as f:
    for item in items:
        f.write(json.dumps(item) + "\n")
```
